```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
datadir = "C:/Users/deber/Documents/TFG_Markel/Fotos herramientas"
categories = ["alicate", "destornillador", "llave_inglesa"]

for category in categories:
    path = os.path.join(datadir, category) #get the directory of each image category
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap = "gray")
        plt.show() #Check the first image example as grayscale
        break
    break
```

```python
#Check the image as an array of 256 bits
print(img_array)
```

```python
#The images have different scales
img_1_D = []
img_2_D = []
for category in categories:
    path = os.path.join(datadir, category) #get the directory of each image category
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        img_1_D.append(img_array.shape[0])
        img_2_D.append(img_array.shape[1])
plt.rcParams["figure.figsize"] = (12,6)
fig, ax = plt.subplots(1,2)
ax[0].hist(img_1_D)
ax[0].set_title("Distribution of vertical pixels in images")
ax[1].hist(img_2_D)
ax[1].set_title("Distribution of horizontal pixels in images")
#Most figures have less than 500 pixels in each dimetion
```
